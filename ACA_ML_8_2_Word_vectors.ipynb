# The 3/4 power subsamples common words

```python
def gaussian_curve(x_vals, mu, sigma):
  scale_factor = 1/np.sqrt(2*np.pi*sigma**2)
  return scale_factor*np.exp(-((x_vals - mu)**2/(2*sigma**2)))
```

```python
x_vals = np.asarray(range(100))
plt.plot(x_vals, gaussian_curve(x_vals, 50, 10), label='1')
plt.plot(x_vals, gaussian_curve(x_vals, 50, 10)**(3/4)/np.sum(
    gaussian_curve(x_vals, 50, 10)**(3/4)), label='3/4')
plt.legend()
plt.show()
```

# Calculate the term-term co-occurrence matrix

```python
dct = Dictionary(corpus)                              # Creates the vocabulary.
bow_corpus = [dct.doc2bow(line) for line in corpus]   # Bag of words.
term_doc_mat = corpus2csc(bow_corpus)                 # Count words in the documents (sentences).
term_term_mat = np.dot(term_doc_mat, term_doc_mat.T)  # term co-occurence matrix (in a sentence).
dct.token2id
```
